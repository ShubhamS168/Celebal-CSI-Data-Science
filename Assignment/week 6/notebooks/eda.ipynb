# Exploratory Data Analysis (EDA)

This notebook provides an exploratory data analysis of the student performance dataset. We will examine the structure of the data, check for missing values, visualize distributions, and understand relationships between key variables and the final grade (G3).

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Load dataset
df = pd.read_csv('../data/student-por.csv', sep=';')
print(df.shape)
df.head()
```

```python

# Check for null values
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python

# Distribution of the target variable (G3)
sns.histplot(df['G3'], kde=True)
plt.title('Distribution of Final Grade (G3)')
plt.show()
```

## Correlation Heatmap

```python

# Correlation matrix for numerical features
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
